**Techniques of Text Processing**

There are four types of text processing techniques, these being: *Topic Analysis; Sentiment Analysis; Intent Detection; and Language Classification*. With some examples of where text processing techniques could be used are in:
- **Topic Analysis:** "*this technique interprets and categorizes large collections of text into topics or themes.*" <br>**For Example:** "*helping to associate currency symbols, numbers, related terms (affordable, pricey) and expressions (it costs a fortune) with the word price*" (DataRobot, 2020).<br><br>
- **Sentiment Analysis:** "*this function automatically detects the emotional undertones of text and classifies them as positive, negative, or neutral.*" <br>**For Example:** "*used in reviews, blogs, forums, and other online media to determine if your customers are happy with their purchases*" (DataRobot, 2020).<br><br>
- **Intent Detection:** "*this classification model detects the intent, purpose, or goal of the text.*" <br>**For Example:** "*to determine whether the intent is to gain information, make a purchase, or unsubscribe from the company*" (DataRobot, 2020).<br><br>
- **Language Classification:** "*this classifies text based on which language itâ€™s written in.*"<br>**For Example:** "*'My Hero Academia' is classified in the English language classification; while 'Shingeki no Kyojin' is classified in the Japanese language classification*" (DataRobot, 2020).<br><br>

##  **Analysis**

In this analysis, we are using Latent Dirichlet Allocation (LDA). "*LDA, which is a popular topic modeling algorithm, is when documents are represented as a mixture of topics and a topic is a bunch of words. Those topics reside within a hidden, also known as a latent layer. LDA would than look at a document to determine a set of topics that are likely to have generated that collection of words. So, if a document uses certain words that are contained in a topic, the document is said to be about that specific topic. Using this topic modeling aglorithm can provide an advantage in our analysis by tuning the LDA parameters to fit different dataset shapes, one can explore topic formation and resulting document clusters*" (*Latent Dirichlet Allocation (LDA)*, MarketMuse). LDA implementation consists of six steps: *loading the data; data cleaning; exploratory data analysis; data pre-processing; LDA model training; and analysing LDA model results.* All of these will be explained in later in the steps for the analysis.

###  **Data Pre-Processing**

In this step we will take the raw data from the dataset and prepare it for our LDA analysis through convertion of the raw data

**Stop Words**

We remove common words in the dataset, known as Stop Words, because these words will be ignored by typical tokenizers and in our analysis

```python
import string # THIS IMPORT ALLOWS US TO SPLIT THE STRING INTO WORDS
import nltk # USED FOR PROCESSING TEXT DATA FROM TEXT-BASED DATASETS
from nltk.corpus import stopwords # THIS IMPORT WILL ALLOW THESE STOP WORDS TO BE IGNORED IN OUR ANALYSIS
nltk.download('stopwords') # THIS DOWNLOADS ALL THE LISTED STOP WORDS FROM THE NLTK PACKAGE
```

```python
# DISPLAY THE LIST OF STOP WORDS
", ".join(stopwords.words("english"))
```

```python
# WE DISPLAY THE TEXT DATA IN LOWER CASE
# DECLARE A VARIABLE FOR LOWER CASE DATA
dataset["text_lower"] = dataset["text"].str.lower()
```

```python
# THIS REMOVES THE STOP WORDS FROM THE TEXT DATA
Stopwords = set(stopwords.words("english"))

def remove_stopwords(text):
    return " ".join([words for words in str(text).split() if words not in Stopwords])

# DECLARE A VARIABLE FOR TEXT DATA WITHOUT STOP WORDS
dataset["text_without_stopwords"] = dataset["text_lower"].apply(lambda text: remove_stopwords(text))
```

**Tokenisation**

This is often used for parsing a large amount of textual data into parts to perform an analysis of the character of the text

```python
#Code Borrowed from:
#Koteeswaren
#2021
#Text Preprocessing in NLP
#Software: Python 
#(Kaggle, link: https://www.kaggle.com/code/beinglegend/text-preprocessing-in-nlp)

from nltk.tokenize import word_tokenize # IMPORT THAT WILL SPLIT DATA INTO SMALLER PORTIONS
nltk.download('punkt') # DOWNLOAD THE TOKEN PACKAGE
```

```python
# THAN WE TOKENISE THE TEXT DATA WITHOUT STOP WORDS
# DECLARE A VARIABLE FOR THE TOKENISED DATA
dataset["tokenisation"] = dataset["text_without_stopwords"].apply(lambda x: word_tokenize(x))
```

**Stemming**

This is the process of producing morphological variants of a root/base word

```python
# NOW WE STEM THE TOKENISED DATA
# DECLARE A VARIABLE FOR STEMMING
dataset['stemming'] = dataset['tokenisation'].apply(lambda x: remove_stopwords(x))
ps = nltk.PorterStemmer()
def stemming(text):
    text = [ps.stem(word) for word in text]
    return text
```

**Lemmatisation**

This is the process of grouping together the different inflected forms of a word so they can be analyzed as a single item

```python
#Code Borrowed from:
#Koteeswaren
#2021
#Text Preprocessing in NLP
#Software: Python 
#(Kaggle, link: https://www.kaggle.com/code/beinglegend/text-preprocessing-in-nlp)

from nltk.stem import WordNetLemmatizer # IMPORT FOR LEMMATISATION
nltk.download('wordnet') # DOWNLOAD THE LEMMATISATION PACKAGE

nltk.download('omw-1.4') # DOWNLOAD THIS PACKAGE
```

```python
# WE LEMMATISE THE STEMMED DATA
# DECLARE A VARIABLE FOR LEMMATISED DATA
dataset['lemminisation'] = dataset['stemming'].apply\
(lambda x: [WordNetLemmatizer().lemmatize(word) for word in x])
```

###  **LDA Model Training**

**Setting up the X and Y**

Here we set up our X and Y, the purpose for this is for future use in splitting the data

```python
# LET X BE TEXT DATA WITHOUT STOP WORDS
X = dataset["text_without_stopwords"]

# LET y BE "user_verified"
y = dataset["user_verified"]
```

**Training and Testing**

We split the dataset into two pieces, one for training and the other for testing, using the X and Y variables we just declared.

```python
from sklearn.model_selection import train_test_split # THIS IMPORT IS USED IN SPLITING THE DATA FOR TESTING AND TRAINING

print(X.shape) # SHOWS THE SIZE OF X
print(y.shape) # SHOWS THE SIZE OF y

# THIS SPLITS THE DATA, test_size=0.3 OR 30%, AND THE TRAIN SIZE IS EQUAL TO THE REMAINING 70%
# WHY random_state=42, IS TO ENSURE THAT THE OUTPUT THAT IS RUN EACH TIME EQUALS TO THE FIRST RUN OUTPUT WHEN SPLITTING
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

print(X_train.shape) # SHOWS THE SIZE OF X_train
print(X_test.shape) # SHOWS THE SIZE OF X_test
print(y_train.shape) # SHOWS THE SIZE OF y_train
print(y_test.shape) # SHOWS THE SIZE OF y_test
```

**Countvectorisation**

This is the process of converting a collection of text documents to a vector of term or token counts

```python
from sklearn.feature_extraction.text import CountVectorizer # IMPORT USED FOR COUNTING THE FREQUENCY OF WORDS

vect = CountVectorizer() # DECLARE THE COUNTVECTOR OBJECT
vect.fit(X_train) # PASS THE THE X_train THROUGH THE COUNTVECTOR OBJECT

# DECLARE A VARIABLE FOR THE TRANSFORMED X_train
X_train_dtm = vect.transform(X_train)

# USE THE DECLARED VARIABLE FOR VECTORISING THE X_train
X_train_dtm = vect.fit_transform(X_train)

print(type(X_train_dtm), X_train_dtm.shape) # DISPLAY THE SHAPE OF X_train_dtm

X_test_dtm = vect.transform(X_test) # DECLARE A VARIABLE FOR THE TRANSFORMED X_test
print(type(X_test_dtm), X_test_dtm.shape) # PRINT THE SIZE
```

**Term Frequencyâ€“Inverse Document Frequency (tf-idf)**

This is the process of quantifying words in a set of documents, by computing a score for each word to signify its importance in the document and corpus

```python
from sklearn.feature_extraction.text import TfidfTransformer # IMPORT FOR TF-IDF

tfidf_transformer = TfidfTransformer() # DECLARE THE TF-IDF OBJECT
tfidf_transformer.fit(X_train_dtm) # PASS THE COUNTVECTORISED X_train_dtm WITH TF-IDF
tfidf_transformer.transform(X_train_dtm) # TRANSFORM THE COUNTVECTORISED X_train_dtm WITH TF-IDF
```

###  **Analysing LDA Model Results**

**Pipeline**

This means performing multiple operations in a sequence, for each element of an iterable, in such a way that the output of each element is the input of the next.

```python
from sklearn.pipeline import Pipeline # IMPORT THE PIPELINE LIBRARY
from sklearn.naive_bayes import MultinomialNB # IMPORT THE GaussianNB LIBRARY

# DECLARE THE PIPELINE OBJECT BY PASSING THROUGH THE CountVectorizer, TfidfTransformer AND MultinomialNB OBJECTS
pipe = Pipeline([
    ('count_vector', CountVectorizer()),
    ('tfidf', TfidfTransformer()),
    ('nb', MultinomialNB())
])
```

```python
# FIT THE PIPELINE WITH X_train & y_train
pipe.fit(X_train, y_train)
```

```python
# THE PIPELINE SCORE SCORES AND PREDICTS THE SPLIT DATA
# THE PIPELINE SCORE HAS A 0.912 ACCURACY
pipe.score(X_test, y_test)
```

```python
# DECLARE A VARIABLE FOR PREDICTIONS
predictions = pipe.predict(X_test)
```

```python
# WE DETERMINE THE ACCURACY OF THE PREDICTIONS
from sklearn import metrics # IMPORT FOR EVALUATING THE MODEL

print("=======Accuracy Score===========") # PRINTS THE ACCURACY SCORE FOR PREDICTIONS
print(metrics.accuracy_score(y_test, predictions))

print("=======Confision Matrix===========") # DISPLAYS THE PREDICTIONS IN A CONFUSION MATRIX
metrics.confusion_matrix(y_test, predictions)
```

```python
# THAN WE GENERATE A CLASSIFICATION REPORT THAT WILL DISPLAY THE PRECISION, RECALL, F1 SCORES
from sklearn.metrics import classification_report

print(classification_report(y_test, predictions)) # WE PASS THROUGH THE y_test AND predictions
```
